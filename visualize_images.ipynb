```python
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
from scripts.train_functions import create_dataloaders
```

```python
dataset_path = 'datasets/exoromper/'
batch = 20
dataloaders, dataset_sizes = create_dataloaders(dataset_path, batch, dataset_name='exoromper')
```

```python
dnames = ['space', 'earth', 'lens_flare']
for dn in dnames:
    print("Sampling from dataset: ",dn)
    dl = dataloaders[dn]
    imgs, lbls, fnames = next(iter(dl))
    imgs_np = imgs.cpu().detach().numpy()

    num_imgs = 3
    # Show inputs into model
    fig, axs = plt.subplots(nrows=1, ncols=num_imgs, figsize=(20,20*num_imgs))
    for i in range(num_imgs):
        img_np_single = imgs_np[i,:]
        axs[i].imshow(np.transpose(img_np_single, axes=[1,2,0]))
        
    # Show original images
    fig, axs = plt.subplots(nrows=1, ncols=num_imgs, figsize=(20,20*num_imgs))
    for i in range(num_imgs):
        f = fnames[i]
        pil_img = Image.open(f).convert('RGB')
        axs[i].imshow(pil_img)
```
